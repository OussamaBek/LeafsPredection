```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import imshow, show
import cv2
import os
from sklearn.svm import SVC
import sklearn.preprocessing as preprocessing
import pandas as pd
import sklearn

print(sklearn.__version__)
```

```python
#extraire les features
extractor = cv2.xfeatures2d.SIFT_create()
def features(image, extractor):
    keypoints, descriptors = extractor.detectAndCompute(image, None)
    return keypoints, descriptors 
def bagsOfFeatures(image):
    key,des = features(image, extractor)
    BOW = cv2.BOWKMeansTrainer(1)
    if des is not None:
        BOW.add(des)# bag of words de l'image
        dictionary = BOW.cluster()
    return dictionary
```

```python
#traitement pour avoir tout les images en parcourant les sous-dossiers du dossier train

folder="./datasetProjetS5_2020/train" #dossier contenant les image du trainning
l=os.listdir('./datasetProjetS5_2020/train') #les sous dossiers du dossier contenant les image du trainning

c=38 # nombre de classe a utilisée
l1=[l[i] for i in range(0,c)]

xtrain = []
ytrain = []

#BOW1 = cv2.BOWKMeansTrainer(c) #bow pour l'histogramme

for label in l1:
    print("chargement pour ", label,end="...")
    for image_path in os.listdir(os.path.join(folder,label)):
        img = cv2.imread(os.path.join(os.path.join(folder,label),image_path))
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)#filtre gray
        key,des = features(gray, extractor)
       #BOW1.add(des)
        dictionary = bagsOfFeatures(gray)#extraire les bags of features de l'image
        if len(dictionary[0]) == 128:#filtre pour les image qui pose problem et qui sont n'egligeable (1 image)
            xtrain.append(dictionary)
            ytrain.append(label)
    print("100%")
```

```python
#convertion en 2d du training
k,r,z=np.array(xtrain).shape
foo_bar = np.array(xtrain).reshape((k, z))
#faire le training sous SVC
clf=SVC()
clf.fit(foo_bar,ytrain)
```

```python
#traitement pour avoir tout les images en parcourant les sous-dossiers du dossier test
#presque le meme que le train
xtest = []
ytest = []

for label in l1:
    print("chargement pour ", label, end="...")
    for image_path in os.listdir("./datasetProjetS5_2020/test/" + label):
        img = cv2.imread("./datasetProjetS5_2020/test/" + label + "/" + image_path)
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        dictionary = bagsOfFeatures(gray)#extraire les bags of features de l'image
        if len(dictionary[0]) == 128:
            xtest.append(dictionary)
            ytest.append(label)
    print("100%")
```

```python
#convertion en 2d
k,r,z=np.array(xtest).shape
foo_bar_test =np.array(xtest).reshape((k, z))
#le score apres le test
s=clf.score(foo_bar_test,ytest)
```
